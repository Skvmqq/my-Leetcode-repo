Problem :- Design an algorithm to convert a binary tree into a string and reconstruct it back to the original tree structure.

Key Idea:
    Use level order traversal (BFS) to store nodes.
    Include "null" for missing children so the exact tree structure can be rebuilt.

Logic:

    Serialize:

        Use a queue for BFS.

        For each node:

        Add its value to the list.

        If the node is None, add "null".

        Join the list into a string.

    Deserialize:

        Split the string into values.

        Create the root node.

        Use a queue to rebuild the tree level by level.

        Assign left and right children accordingly.

Complexity:

    Time: O(n)

    Space: O(n)

```python
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
```

```python
root = TreeNode(1)

root.left = TreeNode(2)
root.right = TreeNode(3)

root.right.left = TreeNode(4)
root.right.right = TreeNode(5)
```

```python
class Codec:

    def serialize(self, root):
        """Encodes a tree to a single string.
        
        :type root: TreeNode
        :rtype: str
        """
        queue = deque([root])
        lis = []

        while queue:
            node = queue.popleft()

            if node:
                lis.append(node.val)
                queue.append(node.left)
                queue.append(node.right)
            else:
                lis.append('null')

        return ','.join(str(v) for v in lis)
                       

    def deserialize(self, data):
        """Decodes your encoded data to tree.
        
        :type data: str
        :rtype: TreeNode
        """
        if not data:
            return None
    
        val=data.split(',')
        if val[0] =='null':
            return None
        root=TreeNode(int(val[0]))
        queue=deque([root])
        i=1
        while queue and i <len(val):
            node=queue.popleft()
            if val[i]!= 'null':
                node.left=TreeNode(int(val[i]))
                queue.append(node.left)
            i+=1
            if i<len(val) and val[i] != 'null':
                node.right=TreeNode(int(val[i]))
                queue.append(node.right)
            i+=1
        return root
```

```python


def print_tree(root):
    queue = deque([root])
    
    while queue:
        node = queue.popleft()
        
        if node:
            print(node.val, end=" ")
            queue.append(node.left)
            queue.append(node.right)
        else:
            print("null", end=" ")
    
    print()
```

```python
codec = Codec()

# serialize
data = codec.serialize(root)
print("Serialized:", data)

# deserialize
new_root = codec.deserialize(data)

print("Tree after deserialize:")
print_tree(new_root)
```
