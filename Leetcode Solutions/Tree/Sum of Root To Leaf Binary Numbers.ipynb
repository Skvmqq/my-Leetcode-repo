

Problem :-
    Each root-to-leaf path represents a binary number.
    Return the sum of all those numbers.

Key Idea:
    While traversing the tree, build the binary number along the path.

Logic:

    Use DFS.

    At each node:

        Left shift current value (current << 1)

        Add the node’s value using bitwise OR

        If the node is a leaf, return the current value.

        Otherwise, return the sum of left and right subtree results.

Complexity:

    Time: O(n)

    Space: O(n) (recursion stack)

Good practice for combining tree traversal with bit manipulation.

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

# Build tree
root = TreeNode(1)

root.left = TreeNode(0)
root.right = TreeNode(1)

root.left.left = TreeNode(0)
root.left.right = TreeNode(1)

root.right.right = TreeNode(1)
```

```python
class Solution:
    def sumRootToLeaf(self, root):
        if root is None:
            return 
        def dfs(node,current):
            if node is None:
                return 0
            current =current<<1|node.val
            if node.left is None and node.right is None:
                return current
            return dfs(node.left,current)+dfs(node.right,current)
            
        return   dfs(root,0)

sol=Solution()
print(sol.sumRootToLeaf(root))
```
