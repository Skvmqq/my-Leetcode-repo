

Problem :-
    Given the root of a Binary Search Tree and an integer k, return the kth smallest element.

Key Idea:
    In a BST, inorder traversal gives elements in sorted order.
    So the kth smallest element is simply the kth element in the inorder traversal.

Approach:
    The first approach that came to my mind was to use inorder traversal, store the values in an array, and then return arr[k-1].

    The second approach: I started thinking about how to remove the array to reduce space.

    The third approach: using iterative inorder traversal.

Logic for third approach:

    Use iterative inorder traversal with a stack.

    Traverse left as much as possible.

    Pop from the stack and increment the count.

    When count equals k, return the value.

    Then move to the right subtree.

Complexity:

    Time: O(h + k) (O(n) in the worst case)

    Space: O(h), where h is the height of the tree



```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


# Build tree
root = TreeNode(5)
root.left = TreeNode(3)
root.right = TreeNode(7)

root.left.left = TreeNode(2)
root.left.right = TreeNode(4)
root.right.right = TreeNode(8)
```

```python
class Solution:
    def kthSmallest(self, root,k):
        # recursive with array (first sol came to my mind) 
        '''arr=[]
        def inorder(node):
            if node is None:
                return
            inorder(node.left)
            arr.append(node.val)
            inorder(node.right)

        inorder(root)
        return arr[k-1]'''
        #recursive without array
        '''count=0
        result=None
        def inorder(node):
            nonlocal count,result
            if not node or result is not None:
                return
            inorder(node.left)
            count+=1
            if count ==k:
                result=node.val
                return     
            inorder(node.right)

        inorder(root)
        return result'''
        if root is None:
            return

        curr = root
        count = 0
        stack = []

        while stack or curr:
            
            while curr:
                stack.append(curr)
                curr = curr.left
            
            curr = stack.pop()
            count += 1
            
            if count == k:
                return curr.val
            
            curr = curr.right

k=3
sol=Solution()
print(sol.kthSmallest(root,k))

```
