

Problem :-
    Given a binary tree, return the number of “good” nodes.
    A node is good if in the path from the root to that node, there are no nodes with a value greater than it.

Key Idea:
    While traversing the tree, keep track of the maximum value seen so far along the path.

Logic:

    Use DFS.

    Pass the current maximum value to each recursive call.

    If node.val >= current_max, increment the count.

    Update the maximum before going deeper.

    Continue for left and right subtrees.

Complexity:

    Time: O(n)

    Space: O(n) (recursion stack)



```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

root = TreeNode(3)
root.left = TreeNode(1)
root.right = TreeNode(4)

root.left.left = TreeNode(3)


root.right.left = TreeNode(1)
root.right.right= TreeNode(5)
```

```python
class Solution:
    def goodNodes(self, root: TreeNode) -> int:
        if root is None:
            return 0

        self.good_count=0
        def dfs(node,current_max):
            if node is None:
                return 0
            if node.val>=current_max:  
                current_max=node.val
                self.good_count+=1   
            dfs(node.left,current_max)
            dfs(node.right,current_max)
        
        current_max=root.val
        dfs(root,current_max)
        return self.good_count

sol=Solution()
print(sol.goodNodes(root))
```
