Problem :-
    Given the root of a binary tree, determine whether it is a valid Binary Search Tree (BST).

Key Idea:
    In a BST, every node must follow the rule:

    Left subtree values < current node

    Right subtree values > current node

    But this condition must hold for the entire subtree, not just the immediate children.

Logic:

    Use recursion with lower and upper bounds.

    If a node’s value is not within (low, high), return False.

    Recursively validate:

    Left subtree with updated upper bound

    Right subtree with updated lower bound

    Start with range (-∞, +∞).

Complexity:

    Time: O(n)

    Space: O(n) (recursion stack)

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

# Build valid BST
root_valid = TreeNode(8)
root_valid.left = TreeNode(3)
root_valid.right = TreeNode(10)

root_valid.left.left = TreeNode(1)
root_valid.left.right = TreeNode(6)
root_valid.right.right = TreeNode(14)

root_valid.left.right.left = TreeNode(4)
root_valid.left.right.right = TreeNode(7)
root_valid.right.right.left = TreeNode(13)
```

```python
class Solution:
    def isValidBST(self, root):
        
        def validate(node,low,high):
            if node is None:
                return True
            if node.val<=low or node.val >= high:
                return False
            return validate(node.left,low,node.val) and validate(node.right,node.val,high)

        return validate(root,float('-inf'),float('inf'))
sol=Solution()
print(sol.isValidBST(root_valid))
```
