Problem :-
    Given a binary tree and a target sum, return the number of paths where the sum of the node values equals the target.
    The path does not need to start at the root, but it must go downward.

Key Idea:
    For every node, count how many paths starting from that node equal the target.
     Then repeat the same process for left and right subtrees.

Logic:

    Write a helper function to count paths starting from a node.

    If current node value equals remaining target, increment count.

    Recursively check left and right with updated target (target - node.val).

    Repeat the process for every node in the tree.

Complexity:

    Time: O(nÂ²) in worst case

    Space: O(n) (recursion stack)

```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


# Build the tree
root = TreeNode(10)
root.left = TreeNode(5)
root.right = TreeNode(-3)

root.left.left = TreeNode(3)
root.left.right = TreeNode(2)
root.right.right = TreeNode(11)

root.left.left.left = TreeNode(3)
root.left.left.right = TreeNode(-2)

root.left.right.right = TreeNode(1)

targetSum=8
```

```python
class Solution:
    def pathSum(self, root, targetSum):

    
        def pathCounter(node,target):
            
            if node is None:
                return 0

            count=0
            if node.val==target:
                count +=1 
        
            count += pathCounter(node.left,target-node.val)
            count += pathCounter(node.right,target-node.val)

            return count
        if root is None:
            return 0

        return (pathCounter(root,targetSum)+self.pathSum(root.left,targetSum)+self.pathSum(root.right,targetSum))


sol = Solution()
print(sol.pathSum(root, targetSum))
```
