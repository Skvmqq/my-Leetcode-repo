

Problem :-
    Given the root of a binary tree, flatten it into a linked list in-place.
    The linked list should follow preorder traversal order.

Key Idea:
    Use reverse preorder traversal (right → left → root).

Why reverse?
    Because we can keep track of the previously processed node and attach it to the current node.

Logic:

    Traverse right subtree first.

    Then traverse left subtree.

    Set current node’s right pointer to previously processed node.

    Set left pointer to None.

    Update the previous node.

    This modifies the tree in-place without using extra data structures.

Complexity:

    Time: O(n)

    Space: O(n) (recursion stack)

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(5)

root.left.left = TreeNode(3)
root.left.right = TreeNode(4)

root.right.right = TreeNode(6)

```

```python
class Solution:
    def flatten(self, root):
        """
        Do not return anything, modify root in-place instead.
        """
        self.prev = None
        
        def dfs(node):
            if not node:
                return

            dfs(node.right)
            dfs(node.left)
            
            node.right = self.prev
            node.left = None
            
            self.prev = node
        
        dfs(root) 


sol=Solution()
sol.flatten(root)
# ---- Print flattened list ----

curr = root
while curr:
    print(curr.val, end=" -> ")
    curr = curr.right
```
