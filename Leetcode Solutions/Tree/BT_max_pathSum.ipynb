```python
Day 65 | Binary Tree Maximum Path Sum 🌳 | DFS

Continuing with Trees — today I solved Binary Tree Maximum Path Sum.

Problem :-
    Given a binary tree, return the maximum path sum.
    A path can start and end at any node, but it must go downward (parent-child connections).

Key Idea:
At each node:

    The maximum path through that node is
    node.val + left + right

    But when returning to the parent, we can only choose one side.

    Also, we ignore negative sums using max(0, subtree_sum).

Logic:

    Use DFS to calculate maximum gain from left and right.

    Ignore negative contributions.

    Update a global maximum at each node.

    Return node.val + max(left, right) to parent.

Complexity:

    Time: O(n)

    Space: O(n) (recursion stack)
```

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

# Build tree
root = TreeNode(-10)
root.left = TreeNode(9)
root.right = TreeNode(20)

root.right.left = TreeNode(15)
root.right.right = TreeNode(7)
```

```python
class Solution:
    def maxPathSum(self, root):
        sum_=float('-inf')
        def maxSum(node):
            nonlocal sum_

            if node is None:
                return 0

            left=max(0,maxSum(node.left))
            right=max(0,maxSum(node.right))
            
            sum_=max(sum_,(node.val +left +right))
            return node.val + max(left,right)

        maxSum(root)
        return sum_
sol=Solution()
print(sol.maxPathSum(root))
```
