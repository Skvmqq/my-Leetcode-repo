Problem :-

    Given an array nums of distinct integers, return all possible permutations.

Key Idea:

    Use backtracking to build permutations by choosing numbers that are not already used in the current path.

Logic:

    Use DFS/backtracking.

    Maintain a path representing the current permutation.

    At each step, try adding numbers that are not already in the path.

    When the path length equals len(nums), store the permutation.

    Backtrack by removing the last element and exploring other possibilities.

Complexity:

    Time: O(n!)

    Space: O(n) (recursion stack)

```python
class Solution:
    def permute(self, nums):

        
        def dfs(path):
            if len(path)==len(nums):
                result.append(path.copy())
                return
            for i in range(0,len(nums)):
                if nums[i] not in path:
                    path.append(nums[i])
                    dfs(path)
                    path.pop()
        
        result=[]
        dfs([])
        return result
    
```

```python
nums = [1,2,3]
sol=Solution()
result=sol.permute(nums)
print(result)
```
