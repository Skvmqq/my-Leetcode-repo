
Problem :-

    Given a collection of numbers that might contain duplicates, return all unique permutations.

Key Idea:

    Use backtracking, but handle duplicates carefully to avoid repeated permutations.

    Sorting the array helps identify duplicates.

Logic:

    Sort the array first.

    Use a used array to track elements already included in the current path.

    Skip duplicates using:
    if i > 0 and nums[i] == nums[i-1] and not used[i-1]

    Build permutations recursively and backtrack.

Complexity:

    Time: O(n! * n)

    Space: O(n) (recursion stack + used array)



```python
class Solution:
    def permuteUnique(self, nums):

        result=[]
        nums.sort()
        used = [False] * len(nums)

        def dfs(path):
            if len(path) == len(nums):
                result.append(path.copy())
                return
            for i in range(len(nums)):
                if used[i]:
                    continue
                if i > 0 and nums[i] == nums[i-1] and not used[i-1]:
                    continue 
                used[i] = True
                path.append(nums[i])
                dfs(path)
                path.pop()
                used[i] = False


        
        dfs([])
        return result
        
```

```python
nums = [1,1,2]
sol=Solution()
sol.permuteUnique(nums)
```
