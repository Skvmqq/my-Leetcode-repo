

Problem :-

    Given an array of distinct integers candidates and a target integer target, return all unique combinations where the chosen numbers sum to target.

    The same number may be chosen multiple times.

Key Idea:

    Use backtracking to explore all possible combinations while keeping track of the running sum.

Logic:

    Use DFS/backtracking.

    Maintain a current path and running total.

    If the total equals the target, store the combination.

    If the total exceeds the target, stop exploring that path.

    Continue exploring by allowing the same element again.

Complexity:

Time: Exponential

Space: O(target)


```python
class Solution:
    def combinationSum(self, candidates, target):
        result = []

        def dfs(start, path, total):
            if total == target:
                result.append(path.copy())
                return

            if total > target:
                return
                
            for i in range(start, len(candidates)):
                path.append(candidates[i])
                dfs(i, path, total + candidates[i]) 
                path.pop()
        
        dfs(0, [], 0)
        return result
        
```

```python
candidates = [2,3,6,7]
target = 7
sol=Solution()
sol.combinationSum(candidates,target)
```
