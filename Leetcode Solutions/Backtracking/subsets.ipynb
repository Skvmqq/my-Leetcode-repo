
Problem :-

    Given an integer array nums, return all possible subsets (the power set).

Key Idea:

    For each element, we have two choices:

    include it in the current subset

    exclude it

    Exploring both choices recursively generates all possible subsets.

Logic:

    Use DFS / backtracking.

    At each index:

    Include the element and continue recursion.

    Backtrack (remove it).

    Explore the path without including the element.

    When we reach the end of the array, store the current subset.

Complexity:

    Time: O(2ⁿ)

    Space: O(n) (recursion stack)

```python
class Solution:
    def subsets(self, nums):
        result = []
        path = []

        def dfs(i):

            if i == len(nums):
                result.append(path[:])
                return
            path.append(nums[i])
            dfs(i + 1)
            path.pop()
            dfs(i + 1)

        dfs(0)
        return result
```

```python
nums = [1,2,3]
sol=Solution()
res=sol.subsets(nums)
print(res)
```
