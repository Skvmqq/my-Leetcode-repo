Problem :-

    Given a string s, partition it such that every substring of the partition is a palindrome.
    Return all possible palindrome partitionings.

Key Idea:

    Use backtracking to try all possible partitions and only continue when the current substring is a palindrome.

Logic:

    Use DFS/backtracking starting from index 0.

    At each step:

        Try all possible substrings starting from current index.

        Check if the substring is a palindrome.

        If yes, add it to the path and continue recursion.

        When the end of the string is reached, store the partition.

        Backtrack and explore other possibilities.

Complexity:

    Time: O(n × 2ⁿ)

    Space: O(n) (recursion stack)

```python
class Solution:
    def partition(self, s):

        result = []

        def dfs(start, path):
            if start == len(s):
                result.append(path.copy())
                return
                
            for end in range(start+1, len(s)+1):
                sub = s[start:end]
                if sub == sub[::-1]:  
                    path.append(sub)
                    dfs(end, path)
                    path.pop()
        
        dfs(0, [])
        return result
        
```

```python
s = "aab"
sol= Solution()
result=sol.partition(s)
print(result)
```
