Problem :-

    Given two integers n and k, return all possible combinations of k numbers chosen from the range 1..n.

Key Idea:

    Use backtracking to build combinations while ensuring numbers are chosen in increasing order to avoid duplicates.

Logic:

    Use DFS/backtracking.

    Maintain a path representing the current combination.

    Start from a given index to avoid reusing previous numbers.

    When the path length equals k, store the combination.

    Backtrack by removing the last element and exploring the next possibilities.

Complexity:

    Time: O(C(n, k))

    Space: O(k) (recursion stack)

```python
class Solution:
    def combine(self, n, k):
        result=[]
        def dfs(start,path):
            if len(path)==k:
                result.append(path.copy())
                return
            for i in range(start,n+1):
                path.append(i)
                dfs(i+1,path)
                path.pop()
            


        dfs(1,[])
        return result
        
```

```python
n= 4 
k=2
sol=Solution()
result=sol.combine(n,k)
print(result)
```
