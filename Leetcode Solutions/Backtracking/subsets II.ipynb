

Problem :-

    Given an integer array nums that may contain duplicates, return all possible subsets without duplicate subsets.

Key Idea:

    Use backtracking to generate subsets, but skip duplicates to avoid repeated subsets.

    Sorting the array helps detect duplicates easily.

Logic:

    Sort the array first.

    Use DFS/backtracking to build subsets.

    At each level:

    Add the current subset to the result.

    Iterate through remaining elements.

    Skip duplicates using
    if i > start and nums[i] == nums[i-1].

Complexity:

    Time: O(2ⁿ)

    Space: O(n) (recursion stack)

```python
class Solution:
    def subsetsWithDup(self, nums):
        nums.sort()
        result = []
        path = []

        def dfs(start):
            result.append(path[:])

            for i in range(start, len(nums)):

                if i > start and nums[i] == nums[i-1]:
                    continue

                path.append(nums[i])
                dfs(i + 1)
                path.pop()

        dfs(0)
        return result
```

```python
nums = [4,4,4,1,4]
sol=Solution()
result=sol.subsetsWithDup(nums)
print(result)
```
