Problem :-

    Given n nodes labeled from 0 to n-1 and a list of edges, determine if the graph forms a valid tree.

Key Idea:

    A graph is a valid tree if:

    It has exactly n - 1 edges
    It has no cycles
    It is fully connected

Logic:

    First check if number of edges is n - 1.
    Build an adjacency list.
    Use DFS to detect cycles:
    Keep track of visited nodes.
    If a visited node is reached again (not parent), a cycle exists.
    After DFS, ensure all nodes are visited (graph is connected).

Complexity:

    Time: O(V + E)
    Space: O(V + E)

```python
class Solution:
    def validTree(self, n, edges):
        if len(edges)!= n-1:
            return False
        adj={}

        for v in range(n):
            adj[v]=[]
        
        for v,u in  edges:
            adj[v].append(u)
            adj[u].append(v)
    
        visited = set()
 
        def dfs(node, parent):
            visited.add(node)

            for neighbor in adj[node]:
                if neighbor not in visited:
                    if dfs(neighbor, node):
                        return True
                elif neighbor != parent:
                    return True  

            return False

    
        if dfs(0, -1):
            return False  

        
        return len(visited) == n
    
            
        
```

```python
n=5
edges = [[0, 1], [0, 2], [0, 3], [1, 4]]
sol =Solution()
print(sol.validTree(n,edges))
```
