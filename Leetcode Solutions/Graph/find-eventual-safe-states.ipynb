Day 74 | Find Eventual Safe States 🌐 | DFS

Starting with Graph problems — today I solved Find Eventual Safe States.

Problem :-
    Given a directed graph, return all nodes that are eventually safe.
    A node is safe if every possible path starting from it leads to a terminal node (a node with no outgoing edges).

Key Idea:
    Use DFS with node states to detect cycles.

    If a node is part of a cycle or leads to a cycle, it cannot be a safe node.

Logic:

    Use three states for each node:

    0 → unvisited

    1 → visiting (currently in DFS path)

    2 → safe node

    If we encounter a node that is already visiting, it means a cycle exists.

    If all neighbors of a node are safe, mark the node as safe.

Complexity:

    Time: O(V + E)

    Space: O(V)



Problem link:

```python
class Solution:
    def eventualSafeNodes(self, graph):
        n = len(graph)
        state = [0] * n  

        def dfs(node):
            if state[node] != 0:
                return state[node] == 2

            state[node] = 1

            for nei in graph[node]:
                if not dfs(nei):
                    return False

            state[node] = 2
            return True

        res = []
        for i in range(n):
            if dfs(i):
                res.append(i)

        return res
```

```python
graph = [[1,2],[2,3],[5],[0],[5],[],[]]
sol=Solution()
lis=sol.eventualSafeNodes(graph)
print(lis)
```
