```python
class Solution:
    def findAnagrams(self, s, p):
        index=[]
        need={}
        for m in p:
            need[m]=need.get(m,0)+1
        freq=len(need)
        left=0
        right=0
        valid=0
        window={}
        while right<len(s):
            ch=s[right]
            right +=1
            if ch in need:
                window[ch]=window.get(ch,0)+1
                if window[ch]==need[ch]:
                    valid +=1
            while right-left>=len(p):
                if valid==freq:
                    index.append(left)
                ch=s[left]
                left +=1
                if ch in need:
                    if window[ch]==need[ch]:
                        valid -=1
                    window[ch] -=1
        return index


s = "cbaebabacd"
p = "abc"
sol=Solution()
indx=sol.findAnagrams(s,p)
print(indx)
```
