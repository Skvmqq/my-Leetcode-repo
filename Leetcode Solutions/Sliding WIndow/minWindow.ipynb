```python
class Solution:
    def minWindow(self, s, t):
        # logic is : first convert text into dictionary
        
        list_=[]
        left =0
        min_len=float('inf')
        subs={}
        right=0
        templ=None
        tempr=None
        if len(t)>len(s):
            return ""

        for i in range(len(t)):
            char=t[i]
            subs[char]=subs.get(char,0)+1
        for right in range(len(s)):
            if s[right] in subs and subs[s[right]] > 0:
                subs[s[right]] -=1
            while all(v==0 for v in subs.values()):
                min_len=min(right-left+1,min_len)
                print(min_len)
                tempr=right
                templ=left
                print(tempr,templ)
                if s[left] in subs :
                    subs[s[left]] +=1
                    left =left+1
                else:
                    left =left+1
        print(min_len)
        return s[templ:tempr]

```
