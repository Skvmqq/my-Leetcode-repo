```python
class Solution:
    def checkInclusion(self, s1, s2):

        need={}
        for s in s1:
            need[s]=need.get(s,0)+1
        window={}
        freq=len(need)
        left=0
        right=0
        valid=0
        while right<len(s2):
            ch=s2[right]
            right +=1
            if ch in need:
                window[ch]=window.get(ch,0)+1
                if window[ch]==need[ch]:
                    valid =valid+1
            while right-left>=len(s1):
                if valid==freq:
                    return True
                ch=s2[left]
                left +=1
                if ch in need:
                    if window[ch]==need[ch]:
                        valid -=1
                    window[ch] -=1
        return False

s1 = "ab"
s2 = "eidbaooo"
sol =Solution()
output=sol.checkInclusion(s1, s2)
print(output)



```
